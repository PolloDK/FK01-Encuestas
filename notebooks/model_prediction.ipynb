```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data/predicciones_diarias.csv")
df_encuestas = pd.read_csv("../data/encuestas.csv")
```

```python
df["date"] = pd.to_datetime(df["date"])
df_encuestas["date"] = pd.to_datetime(df_encuestas["date"])
```

```python
df.head()
```

```python
df_features = pd.read_csv("../data/features_dataset.csv")
df_features["date"] = pd.to_datetime(df_features["date"])
fecha_corte = df_features[df_features["in_sample"] == False]["date"].min()
```

```python
plt.figure(figsize=(15, 5))
plt.plot(df_encuestas['date'], df_encuestas['aprobacion_boric'], linestyle='-', color='blue', label='Aprobación Real')
plt.plot(df['date'], df['prediccion_aprobacion'], linestyle='--', color='orange', label='Predicción')
plt.axvline(x=fecha_corte, color="black", linestyle="-", label=f"Corte entrenamiento: {fecha_corte.date()}")
plt.title('Aprobación de Boric - Real vs. Predicción (Random Forest)')
plt.xlabel('Fecha')
plt.ylabel('Aprobación (%)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
